```python
#9.6. 
#   ,   ? 
#  ,   .

pie_df = churn_df.copy()
pie_df['count'] = 1
pie_df['Gender exited'] = pie_df[
    (pie_df['Exited']==1) & ((pie_df['Gender']=='Male') | (pie_df['Gender']=='Female'))
    ]['Gender']

pie_df = pie_df.groupby(
    by=['Gender exited','Gender'],
    as_index=False
)['count'].count()
#display(pie_df)

fig = px.pie(
    pie_df,
    values='count', 
    names='Gender exited',
    title='Male and female exited customers',       
        
)
fig.show();

'''
:
     .
'''
```

```python
#9.7. 
#          ? (Exited/NumOfProducts)
#         . (bar)

bar_df = churn_df.copy()
bar_df['count'] = 1

bar_df = bar_df.groupby(
    by=['Exited','NumOfProducts'],
    as_index=False
)['count'].count()
bar_df['rates'] = None

Loyal_sum = bar_df[(bar_df['Exited']==0)]['count'].sum()
Exited_sum = bar_df[(bar_df['Exited']==1)]['count'].sum()
#print('Loyal_sum = ',Loyal_sum,'Exited_sum = ',Exited_sum)
#display(bar_df)

for i in range(bar_df.shape[0]):
    if bar_df['Exited'].iloc[i]==0:          
        bar_df['rates'].iloc[i] = bar_df['count'].iloc[i]/Loyal_sum*100   
    if bar_df['Exited'].iloc[i]==1:            
        bar_df['rates'].iloc[i] = bar_df['count'].iloc[i]/Exited_sum*100    
#display(bar_df)
    
bar_df['Exited'] = bar_df['Exited'].astype('category')
bar_df['Exited'] = bar_df['Exited'].cat.rename_categories({1:'Exited',0:'Loyal'})

fig = px.bar(
    bar_df, 
    x='Exited',
    y='rates',
    color='NumOfProducts',
    title='Products of exited and loyal customers'        
)
fig.show();

'''
:
    1,2  3  (46%, 53%  1% ).
    1,2, 3   4  (69%, 17%, 11%  3% ).
    ,        ,
 ,      .
'''
```

```python
#9.8. 
#         ?  (IsActiveMember/Exited)
#  ,  .     ,      ?

bar_df = churn_df.copy()
bar_df['count'] = 1


bar_df = bar_df.groupby(
    by=['Exited','IsActiveMember'],
    as_index=False
)['count'].count()
#display(bar_df)
bar_df['rates'] = None

Loyal_sum = bar_df[(bar_df['Exited']==0)]['count'].sum()
Exited_sum = bar_df[(bar_df['Exited']==1)]['count'].sum()
#print('Loyal_sum = ',Loyal_sum,'Exited_sum = ',Exited_sum)
#display(bar_df)

for i in range(bar_df.shape[0]):
    if bar_df['Exited'].iloc[i]==0:          
        bar_df['rates'].iloc[i] = bar_df['count'].iloc[i]/Loyal_sum*100   
    if bar_df['Exited'].iloc[i]==1:            
        bar_df['rates'].iloc[i] = bar_df['count'].iloc[i]/Exited_sum*100    
#display(bar_df)
    
bar_df['Exited'] = bar_df['Exited'].astype('category')
bar_df['Exited'] = bar_df['Exited'].cat.rename_categories({1:'Exited',0:'Loyal'})

fig = px.bar(
    bar_df, 
    x='Exited',
    y='rates',
    color='IsActiveMember',    
    title='Products of exited and loyal customers',
    color_continuous_scale=[(0.0, "blue"),   (1.0, "green")],      
)
fig.show();

'''
:
   55%  .
    36%.
    ,     ,
      .
  64%     ,   
      ,    
  (  , 
  ,   
    . .)
'''
```

```python
#9.9. 
#       ?   ,       . 
# ,      .

chor_df = churn_df.copy()
chor_df['count'] = 1
#display(chor_df)

chor_df = chor_df.groupby(
    by=['Exited','Geography'],
    as_index=False
)['count'].count()
#display(chor_df)

data_df = chor_df[chor_df['Exited']==1]
data_df['rates'] = None
#display(data_df)

Exited_sum = data_df[(chor_df['Exited']==1)]['count'].sum()
#print('Loyal_sum = ',Loyal_sum,'Exited_sum = ',Exited_sum)
#display(chor_df)


for i in range(data_df.shape[0]):    
    if data_df['Exited'].iloc[i]==1:                     
        data_df['rates'].iloc[i] = data_df['count'].iloc[i]/Exited_sum*100  
        
#display(data_df)
    


fig = px.choropleth(
    data_frame=data_df, #DataFrame
    locations="Geography", #  
    locationmode = "country names", #     Plotly
    color="rates", #   
    hover_name="Exited", #       
    title='Global total_vaccinations of COVID-19', #
    width=800, #
    height=500, #
    color_continuous_scale='Reds' # 
)
fig.show();

'''
:
     39.96%, 39.76%  20.28%
 ,    .
  ,         ,
     .
     
'''
```
