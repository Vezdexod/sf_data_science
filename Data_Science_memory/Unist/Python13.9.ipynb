```python
import pandas as pd
import plotly
import plotly.express as px
print(plotly.__version__)
```

```python
churn_df = pd.read_csv('C:/Python/PYTHON-13/churn.csv', sep=',')
churn_df = churn_df.drop(['RowNumber'], axis=1)
#display(churn_df.head())
```

```python
#9.1. 
# Каково соотношение ушедших и лояльных клиентов? 
# Покажите это на графике и дайте комментарий по соотношению.
# Подходит круговая диаграмма

pie_df = churn_df.copy()
pie_df['count'] = 1
pie_df['Exited'] = pie_df['Exited'].astype('category')
pie_df['Exited'] = pie_df['Exited'].cat.rename_categories({1:'Exited',0:'Loyal'})

fig = px.pie(
    pie_df,
    values='count', 
    names='Exited', 
    title='Exited and loyal customers',       
        
)
fig.show();

'''
Вывод:
Количество ушедших клиентов составляет 20,4%.
Можем сделать вывод, что каждый 2 из 10 клиентов отказался от услуг данного банка.
'''
```

```python
#9.6. 
# Кто чаще уходит, мужчины или женщины? 
# Постройте график, который иллюстрирует это.

pie_df = churn_df.copy()
pie_df['count'] = 1
pie_df['Gender exited'] = pie_df[
    (pie_df['Exited']==1) & ((pie_df['Gender']=='Male') | (pie_df['Gender']=='Female'))
    ]['Gender']

pie_df = pie_df.groupby(
    by=['Gender exited','Gender'],
    as_index=False
)['count'].count()
#display(pie_df)

fig = px.pie(
    pie_df,
    values='count', 
    names='Gender exited',
    title='Male and female exited customers',       
        
)
fig.show();

'''
Вывод:
Чаще из данного банка уходят женщины.
'''
```

```python
#9.7. 
# Как отток клиентов зависит от числа приобретённых у банка услуг? (Exited/NumOfProducts)
# Для ответа на этот вопрос постройте многоуровневую столбчатую диаграмму. (bar)

bar_df = churn_df.copy()
bar_df['count'] = 1

bar_df = bar_df.groupby(
    by=['Exited','NumOfProducts'],
    as_index=False
)['count'].count()
bar_df['rates'] = None

Loyal_sum = bar_df[(bar_df['Exited']==0)]['count'].sum()
Exited_sum = bar_df[(bar_df['Exited']==1)]['count'].sum()
#print('Loyal_sum = ',Loyal_sum,'Exited_sum = ',Exited_sum)
#display(bar_df)

for i in range(bar_df.shape[0]):
    if bar_df['Exited'].iloc[i]==0:          
        bar_df['rates'].iloc[i] = bar_df['count'].iloc[i]/Loyal_sum*100   
    if bar_df['Exited'].iloc[i]==1:            
        bar_df['rates'].iloc[i] = bar_df['count'].iloc[i]/Exited_sum*100    
#display(bar_df)
    
bar_df['Exited'] = bar_df['Exited'].astype('category')
bar_df['Exited'] = bar_df['Exited'].cat.rename_categories({1:'Exited',0:'Loyal'})

fig = px.bar(
    bar_df, 
    x='Exited',
    y='rates',
    color='NumOfProducts',
    title='Products of exited and loyal customers'        
)
fig.show();

'''
Вывод:
У лояльных клиентов имеется 1,2 и 3 продукта (46%, 53% и 1% соответственно).
У ушедших клиентов было 1,2, 3 и даже 4 продукта (69%, 17%, 11% и 3% соответственно).
Из этого можно сделать вывод, что ушедшие клиенты являлись болле активными клиентами банка,
чем лояльные, так как пользовались большим числом продуктов.
'''
```

```python
#9.8. 
# Как влияет наличие статуса активного клиента на отток клиентов?  (IsActiveMember/Exited)
# Постройте диаграмму, иллюстрирующую это. Что бы вы предложили банку, чтобы уменьшить отток клиентов среди неактивных?

bar_df = churn_df.copy()
bar_df['count'] = 1


bar_df = bar_df.groupby(
    by=['Exited','IsActiveMember'],
    as_index=False
)['count'].count()
#display(bar_df)
bar_df['rates'] = None

Loyal_sum = bar_df[(bar_df['Exited']==0)]['count'].sum()
Exited_sum = bar_df[(bar_df['Exited']==1)]['count'].sum()
#print('Loyal_sum = ',Loyal_sum,'Exited_sum = ',Exited_sum)
#display(bar_df)

for i in range(bar_df.shape[0]):
    if bar_df['Exited'].iloc[i]==0:          
        bar_df['rates'].iloc[i] = bar_df['count'].iloc[i]/Loyal_sum*100   
    if bar_df['Exited'].iloc[i]==1:            
        bar_df['rates'].iloc[i] = bar_df['count'].iloc[i]/Exited_sum*100    
#display(bar_df)
    
bar_df['Exited'] = bar_df['Exited'].astype('category')
bar_df['Exited'] = bar_df['Exited'].cat.rename_categories({1:'Exited',0:'Loyal'})

fig = px.bar(
    bar_df, 
    x='Exited',
    y='rates',
    color='IsActiveMember',    
    title='Products of exited and loyal customers',
    color_continuous_scale=[(0.0, "blue"),   (1.0, "green")],      
)
fig.show();

'''
Вывод:
Среди лояльных клиентов 55% являются активными.
Среди ушедших активными являлись 36%.
Из этого можно сделать вывод, что чем выше активность клиента,
тем ниже риск его ухода из банка.
Так как 64% ушедших клиентов являлись не активными, нужно было 
больше вовлекать их в сотрудничесво с банком, предлагать бонусы и 
делать подарки (например повышенный кэшбэк, 
бесплатное обслуживание карты, увеличенный беспроцентный период
по кредитной карте и т. п.)
'''
```

```python
#9.9. 
# В какой стране доля ушедших клиентов больше? Постройте тепловую картограмму, которая покажет это соотношение на карте мира. 
# Предположите, с чем это может быть связано.

chor_df = churn_df.copy()
chor_df['count'] = 1
#display(chor_df)

chor_df = chor_df.groupby(
    by=['Exited','Geography'],
    as_index=False
)['count'].count()
#display(chor_df)

data_df = chor_df[chor_df['Exited']==1]
data_df['rates'] = None
#display(data_df)

Exited_sum = data_df[(chor_df['Exited']==1)]['count'].sum()
#print('Loyal_sum = ',Loyal_sum,'Exited_sum = ',Exited_sum)
#display(chor_df)


for i in range(data_df.shape[0]):    
    if data_df['Exited'].iloc[i]==1:                     
        data_df['rates'].iloc[i] = data_df['count'].iloc[i]/Exited_sum*100  
        
#display(data_df)
    


fig = px.choropleth(
    data_frame=data_df, #DataFrame
    locations="Geography", #столбец с локациями
    locationmode = "country names", #режим сопоставления локаций с базой Plotly
    color="rates", #от чего зависит цвет
    hover_name="Exited", #группирующая переменная      
    title='Global total_vaccinations of COVID-19', #заголовок
    width=800, #ширина
    height=500, #высота
    color_continuous_scale='Reds' #палитра цветов
)
fig.show();

'''
Вывод:
Отношение числа шедших клиентов равны 39.96%, 39.76% и 20.28%
для Германии, Франции и Испании соответственно.
Из этого следует, что хоть Германия и лидирует по числу ушедших клиентов,
но их отношение поти равно Франции.
Скорее всего это связано с ДОПИСАТЬ
'''
```
